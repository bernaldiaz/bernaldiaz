```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Crear datos para la curva sinusoidal
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Configurar la figura y el eje
fig, ax = plt.subplots()
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-1, 1)
line, = ax.plot([], [], 'o-', lw=2)

# Función de inicialización de la animación
def init():
    line.set_data([], [])
    return line,

# Función de animación para actualizar los datos en cada cuadro
def animate(i):
    x_data = x[:i]
    y_data = y[:i]
    line.set_data(x_data, y_data)
    return line,

# Crear la animación
ani = FuncAnimation(fig, animate, frames=len(x), init_func=init, blit=True, interval=50)

# Convertir la animación a HTML para mostrarla en Colab
HTML(ani.to_html5_video())
```
