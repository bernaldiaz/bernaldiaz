```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def caida_libre(tiempo_total, intervalo_tiempo, gravedad=9.8):
    tiempos = np.arange(0, tiempo_total + intervalo_tiempo, intervalo_tiempo)
    espacios = 0.5 * gravedad * tiempos**2
    velocidades = gravedad * tiempos

    return tiempos, espacios, velocidades
```

```python
def graficar_caida_libre(tiempos, espacios, velocidades):
    plt.figure(figsize=(10, 6))

    # Graficar espacio-tiempo
    plt.subplot(2, 1, 1)
    plt.plot(tiempos, espacios, label='Espacio')
    plt.title('Caida Libre - Espacio-Tiempo')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Espacio (m)')
    plt.legend()

  # Graficar velocidad-tiempo
    plt.subplot(2, 1, 2)
    plt.plot(tiempos, velocidades, label='Velocidad', color='orange')
    plt.title('Caida Libre - Velocidad-Tiempo')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Velocidad (m/s)')
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
if __name__ == "__main__":
    tiempo_total = 10  # segundos
    intervalo_tiempo = 0.1  # segundos

    tiempos, espacios, velocidades = caida_libre(tiempo_total, intervalo_tiempo)
    graficar_caida_libre(tiempos, espacios, velocidades)
```
