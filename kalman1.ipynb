```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# Definir parmetros
A = np.eye(2)
Bw = np.eye(2)
C = np.eye(2)
D = np.zeros((2, 2))
maxT = 1000  # Nmero de pasos de simulacin

# Generar entradas aleatorias
w = 0.1 * np.random.randn(2, maxT)  # Entrada aleatoria
v = 0.01 * np.random.randn(2, maxT)  # Ruido

# Definir el sistema NCP
ncp = signal.dlti(A, Bw, C, D, dt=1)  # Especificamos dt=1 para discreto

# Simular la respuesta del sistema
_, _, z = signal.dlsim(ncp, w.T)  # Transponer w para ajuste correcto

# Agregar ruido
z += v.T

# Graficar los resultados
plt.figure(figsize=(6, 6))
plt.plot(z[:, 0], z[:, 1], label='Simulacin NCP')
plt.xlabel('z1')
plt.ylabel('z2')
plt.title('Simulacin del modelo NCP')
plt.legend()
plt.grid()
plt.show()
```
