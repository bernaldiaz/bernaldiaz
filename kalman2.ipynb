```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal

# Definir parmetros
A = np.eye(2)
Bw = np.eye(2)
C = np.eye(2)
D = np.zeros((2, 2))
maxT = 1000  # Nmero de pasos de simulacin

# Generar entradas aleatorias
w = 0.1 * np.random.randn(2, maxT)  # Entrada aleatoria
v = 0.01 * np.random.randn(2, maxT)  # Ruido
x = np.zeros((2, maxT))  # Almacenamiento para x

# Corregir inicializacin incorrecta de x[:,1]
x[:, 0] = [0, 0]  # Inicializar el primer estado en lugar del segundo

# Corregir la actualizacin de x en el bucle
for k in range(1, maxT):  # Comenzar en 1 en lugar de 2
    x[:, k] = A @ x[:, k - 1] + w[:, k - 1]  # Usar @ para multiplicacin matricial

# Calcular salida z correctamente
z = C @ x + v  # Usar @ para multiplicacin matricial

# Graficar los resultados
plt.figure(figsize=(6, 6))
plt.plot(z[0, :], z[1, :], label='Simulacin NCP')
plt.xlabel('z1')
plt.ylabel('z2')
plt.title('Simulacin del modelo NCP')
plt.legend()
plt.grid()
plt.show()
```
