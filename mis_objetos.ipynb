```python
import random
```

```python
class Rectangulo():
  def __init__(self, base = 1, altura = 2, color = 'verde'):
    self.base = base
    self.altura = altura
    self.color = color

  @property
  def perimetro(self):
    return 2 * self.altura * self.base

  def __str__(self):
    return f'el rectangulo tiene una base {self.base} y \nuna altura {self.altura}'

  @staticmethod
  def soniguales(rect1, rect2):
    if rect1.base == rect2.base and rect1.altura == rect2.altura:
      return True
    else:
      return False

  @classmethod
  def rectangulo_aleatorio(cls):
    base = random.randrange(1,10)
    altura =random.randrange(1,10)
    return cls(base,altura)
```

```python
rectangulo1 = Rectangulo(5,6,'azul')

print(f'el perimetro del rectangulo 1 es {rectangulo1.perimetro()}')
```

```python
print(rectangulo1)
```

```python
rect1 = Rectangulo(3,5)
rect2 = Rectangulo(7-2,5+3)
print(rect1, "\n")
print(rect2, "\n")
print(Rectangulo.soniguales(rect1,rect2))
```

```python
rect3 = Rectangulo.rectangulo_aleatorio()
print(rect3)
```
